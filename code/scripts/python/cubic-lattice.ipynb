```python
import kwant
from matplotlib import pyplot as plt
import numpy as np
import random 
```

```python
def make_square_lattice(W, a=1, t=1.0, Nx=50, Ny=20):
    
    lat = kwant.lattice.cubic(a)
    
    syst = kwant.Builder()
    
    syst[(lat(x,y,z) for x in range(Nx) for y in range(Nx) for z in range(Nx))] = -4*t + random.uniform(-W/2, W/2)
    
    syst[lat.neighbors()] = -t 
    
    return syst
```

```python
s = make_square_lattice(20)

#kwant.plot(s)
```

```python
sp = kwant.kpm.SpectralDensity(s.finalized())
```

```python
e,s = sp()
```

```python
plt.plot(e,s)
```
