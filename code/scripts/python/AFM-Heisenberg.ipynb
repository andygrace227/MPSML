# AFM XYZ Model 
 The model

$$ H = -\frac{1}{2} \sum_{i=0}^{N-1} \left((J+\lambda) \sigma_i^x \sigma_{i+1}^x + (J-\lambda) \sigma_i^y \sigma_{i+1}^y  + \Delta \sigma_i^z \sigma_{i+1}^z + (-1)^ih^z_i \sigma_i^z \right) $$

The small site dependent $h$ field is used to remove the degeneracy of the ground state. We can do so by setting $h_i = (-1)^ih$

The operators are understood to be the $N$-fold product of the spin space

$$ \sigma_i = \mathbf{1} \otimes ... \otimes \sigma \otimes \mathbf{1} ... \otimes \mathbf{1} $$

```python
import matplotlib.pyplot as plt
import numpy as np
from random import seed, uniform
from qutip import *
import sys
import os
from scipy.interpolate import interp1d
```

### Define the functions to construct the energy operator

```python
def transverse(N, site):
    op = []
    
    for i in range(N):
        op.append(qeye(2))
        
    op[site] = sigmax()
    
    return tensor(op)

def longitudinal(N, site):
    op = []
    
    for i in range(N):
        op.append(qeye(2))
        
    op[site] = sigmaz()
    
    return tensor(op)

def surface(N, site):
    op = []
    
    for i in range(N):
        op.append(qeye(2))
        
    op[site] = sigmay()
    
    return tensor(op)

def x_term(N, site):
    if (site == N-1):
        return transverse(N,site) * transverse(N, 0) 
    else:
        return transverse(N,site) * transverse(N, site+1)
    
def y_term(N, site):
    if (site == N-1):
        return surface(N,site) * surface(N,0) 
    else:
        return surface(N,site) * surface(N, site+1)
    
def z_term(N, site):
    if (site == N-1):
        return longitudinal(N,site) * longitudinal(N, 0)
    else:
        return longitudinal(N,site) * longitudinal(N, site+1)
    
def H(N, J, lam, delta, h):
    term1 = x_term(N, 0) 
    term2 = y_term(N, 0)
    term3 = z_term(N, 0)
    term4 = longitudinal(N,0)
       
    for site in range(1, N):
        term1 += x_term(N,site)
        term2 += y_term(N,site)
        term3 += z_term(N,site)
        term4 += (-1)**site * longitudinal(N,site)
        
    Jx = J + lam
    Jy = J - lam
    Jz = delta
    
    return -0.5/N * (Jx * term1 + Jy * term2 + Jz*term3 + h*term4)

def mag(N):
    tmp = longitudinal(N, 0)
    for site in range(1, N):
        tmp += (-1)**site * longitudinal(N,site)
    return tmp / (2 * N)
```

```python
def f(x):
    return x - 1

x1 = [-0.61, -0.65, -0.67, -0.56, -0.43, -0.22, -0.08, 0.12, 0.45, 0.86, 1.14, 1.35, 1.45, 1.52, 1.50]
y1 = [-0.54, -0.35, -0.16, -0.11,  0.32,  0.55,  0.67, 0.82, 1.07, 1.15, 1.06, 0.85, 0.67, 0.42, 0.23]
g = interp1d(x1,y1)

x = np.linspace(-2,2,20)
xn = np.linspace(-0.61,1.5,1500)
y = f(x)

plt.plot(x,y)
plt.plot(-x,y)
plt.plot(x,-y)
plt.plot(-x,-y)
plt.plot(xn, g(xn))
plt.plot(x,[0 for i in range(len(x))])
plt.ylim([-1.5,1.5])
plt.show()
```

```python
plt.spy(H(3,0.5,.2,0.9, 0.05).full())
plt.show()
```

```python
data = []
J = -1
for i in x1:
    delta = i
    lam = g(delta)
    data.append(H(3, -1, lam, delta, 0.01).groundstate())
```

```python
vals = []
vecs = []
for i in range(len(data)):
    vals.append(data[i][0])
    if data[i][1][1][0][0].real < 0:
        vecs.append(-data[i][1])
    else:
        vecs.append(data[i][1])
```

```python
interps = []
for i in range(8): 
    n = i
    y = [vecs[i][n][0][0].real for i in range(len(vecs))]
    x = [i for i in range(len(y))]
    plt.plot(x, y, label="Exact")
    c = interp1d(x, y, kind='linear')
    interps.append(c)
    xnew = np.linspace(0,len(vecs)-1,1000)
    plt.ylabel(f"c_{i}")
    plt.xlabel("Point on RG flow diagram")
    plt.plot(xnew, c(xnew), linestyle='--', color='red', label="Linear Interpolant")
    plt.legend()
    plt.savefig(f"c_{i}")
    plt.show()
```

```python
def interpolated_wavefx(point):
    t = []
    for i in range(8):
        t.append(interps[i](point))
    
    return Qobj(np.array(t))
```

```python
m_targ = []
for i in range(len(data)):
    m_targ.append((data[i][1].dag() * mag(3) * data[i][1]).full()[0][0].real)
```

```python
m_pred = []
for i in np.linspace(0,14,1500):
    m_pred.append(np.dot(interpolated_wavefx(i).full().real.T , np.dot(mag(3).full().real , interpolated_wavefx(i).full().real))[0][0])
```

```python
plt.plot(m_targ, label="Exact Diagonalization")
plt.plot(m_pred, label="From Interpolated Coefficients", linestyle='--', color='red')
plt.legend(loc='upper right')
plt.ylabel("Staggerd Magnetization")
plt.xlabel("Point on RG flow diagram")
plt.savefig("magnetization")
plt.title("Anisotropic Heisenberg XYZ Magnetization")
plt.show()
```
